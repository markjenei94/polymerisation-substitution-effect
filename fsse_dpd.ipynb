```python
import os

import matplotlib.pyplot as plt

from fsse import FSSE, ReadReactions

conversion_dict = {}
fsse_dict = {}
norm_dict = {}
raw_norm_dict = {}

PATH = os.getcwd()
```

### DGEBA-P4A

```python
data_dir = PATH + "/data/DGEBA_P4A/dpd_outfiles/"
monomer_keys = data_dir + "/monomer_keys.txt"
reaction_names = ["Primary", "Secondary"]
rr = ReadReactions("DPD", data_dir, "", [], n_runs=10)

fsse = FSSE(loads=[2000, 1000], funcs=[2, 4], reactions=rr.reactions)

name = "dgeba-p4a"
norm_dict[name], raw_norm_dict[name] = fsse.norm(monomer_type_idx=1, conv_cutoff=0.48, smooth_factors=[0.2, 0.1],
                                                 plot=True)

conversion_dict[name], fsse_dict[name] = fsse.fsse(monomer_type_idx=1, smooth_method="loess",
                                                   smooth_factors=[0.2, 0.25, 0.3, 0.5], norm=norm_dict[name])

for i in range(len(conversion_dict[name])):
    plt.plot(conversion_dict[name][i], fsse_dict[name][i])
plt.ylim([0.5, 4.5])
plt.grid(axis='y')
```

```python
conversion_dict[name + "_savgol"], fsse_dict[name + "_savgol"] = fsse.fsse(monomer_type_idx=1, smooth_method="savgol",
                                                                           smooth_factors=[0.2, 601, 501, 1001],
                                                                           norm=norm_dict[name],
                                                                           savgol_deg=[3, 2, 1, 1])

for i in range(len(conversion_dict[name + "_savgol"])):
    plt.plot(conversion_dict[name + "_savgol"][i], fsse_dict[name + "_savgol"][i])
plt.ylim([0.5, 6.5])
plt.grid(axis='y')
```
